```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("smartwatches.csv")
print("DataFrame loaded from CSV file:")
print(df.to_string())
```

**Summary of the data**

```python
print("\nSummary of sells of smart watches:")
print(df.info())
```

**1. Display the products with more than 50% discount.**

```python
print("Products having 50% or more discount : ")
print(df[df["Discount Percentage"]>=50.0].to_string())
```

**2. Display the products with rating over 4**

```python
print("The products with above 4* rating:")
print(df[df["Rating"]>=4.0].to_string())
```

**3. Display the products with more than 1000 reviews.**

```python
print("Products with more than 1000 reviews:")
print(df[df["Number OF Ratings"]>=1000].to_string())
```

**4. Display the Fire Boltt products with more than 2999 current price.**

```python
print("Products by Fire Boltt with more than 2999 current price :")
print(df[df["Brand"]=='fire-boltt'][df["Current Price"]>2999].to_string())
```

**5. Products under 3000 with more than 4 star rating**

```python
print("Products under 3000 :")
print(df[df["Current Price"]<3000][df["Rating"]>=4.0].to_string())
```

**6. Display the products with less than 3 star rating**

```python
print("Products with less than 3* rating :")
print(df[df["Rating"]<3.0].to_string())
```

**7. Products with circular dial shape.**

```python
print("Products with Circular dial :")
print(df[df["Dial Shape"]=='Circle'].to_string())
```

**8. Display the products with 1.4 inches display size.**

```python
print("Products with 1.4 inches display:")
print(df[df["Display Size"]=='1.4 inches'].to_string())
```

**9. Display the products by boAt**

```python
print("Products by boAt :")
print(df[df["Brand"]=='boat'].to_string())
```

**10. Display the products with silicon straps.**

```python
print("Products with Silicon straps :")
print(df[df["Strap Material"]=='Silicon'].to_string())
```
